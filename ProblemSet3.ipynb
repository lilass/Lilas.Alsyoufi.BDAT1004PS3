```python
import pandas as pd #import pandas
marital = pd.read_csv('us-marriages-divorces-1867-2014.csv') #read csv into dataframe 
marital.head()
```

```python
marital.info()
```

```python
#fix the null values in the set to hide the fill in the gaps in hte chart. filling with zero or average didn't make sense
#since the number of marriages and divorces doesn't change a lot each year, I used ffill to fill the missing values with the previous column value
fullMarital= marital.ffill(axis=0)
```

```python
fullMarital.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x_data = fullMarital['Year']
# create x data


new_marital = fullMarital
new_marital['divPerCapita']=new_marital['Divorces'] / new_marital['Population']
new_marital['marPerCapita']=new_marital['Marriages'] / new_marital['Population']


columns = new_marital.columns.drop(['Year','Marriages','Divorces','Population','Marriages_per_1000','Divorces_per_1000'],1)
#set columns values 


# create figure and axis
fig, ax = plt.subplots()
plt.xlabel("Year")
plt.ylabel("Average per Capita")


# plot each column
ax.plot(x_data, new_marital['divPerCapita'],label='Divorces')
ax.plot(x_data, new_marital['marPerCapita'], label = 'Marriages')
# set title and legend
ax.set_title('US Marriages and Divorces Per Capita 1867 - 2014 ')
ax.legend()
```
